```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import f1_score, auc, roc_curve


titanic = sns.load_dataset('titanic')
titanic.isnull().sum()


X = titanic[['pclass','sex','age','sibsp','parch','fare']]
Y = titanic['survived']

imputer = SimpleImputer(strategy = 'mean')
X['age'] = imputer.fit_transform(X[['age']])
X['fare'] = imputer.fit_transform(X[['fare']])
X['sex'] = LabelEncoder().fit_transform(X['sex'])

scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns = X.columns)
X_train, X_test, y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
model = AdaBoostClassifier(n_estimators = 100,learning_rate=0.05, random_state = 42)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
f1 = f1_score(Y_test,y_pred)
print(f"f1 score : {f1}")

scores = cross_val_score(model, X_test, Y_test, cv=5, scoring='accuracy')  # cv=5는 5-fold 교차 검증
print(f"교차 검증 정확도: {scores}")
print(f"평균 정확도: {np.mean(scores)}")
```

```python
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.impute import SimpleImputer
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
data = pd.read_csv(url)
```

```python
X = data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Embarked']]
y = data['Fare']
```

```python
imputer = SimpleImputer(strategy='mean')
X['Age'] = imputer.fit_transform(X[['Age']])
```

```python
X['Embarked'].fillna('S', inplace=True)
X['Sex'] = LabelEncoder().fit_transform(X['Sex'])
X['Embarked'] = LabelEncoder().fit_transform(X['Embarked'])
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X_train, X_test , Y_train, Y_test = train_test_split(X, y, test_size = 0.2)
```

```python
model = RandomForestRegressor(n_estimators = 100)
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(Y_test, y_pred)
print("Mean Squared Error (MSE):", mse)
```
